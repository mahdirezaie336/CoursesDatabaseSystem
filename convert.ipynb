```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./students.csv')
df
```

```python
df.birth_date = df.birth_date.str.replace('/', '-')
df
```

```python
def split_name(row):
    names = row['name_en'].split(' - ')
    row['first_name_en'] = names[0]
    row['last_name_en'] = names[1]
    return row
```

```python
df = df.apply(split_name, axis=1)
```

```python
df.to_csv('./students-modified.csv', index=False)
```

```python
df = pd.read_csv('./faculty.csv')
df.birth_date = df.birth_date.str.replace('/', '-')
df = df.apply(split_name, axis=1)
df.to_csv('./faculty-modified.csv', index=False)
```
